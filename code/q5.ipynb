```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(1111111)

N = [5, 10, 20, 40, 60, 80, 100, 500, 10**3 , 10**4]
M = 100
```

### Generating Dataset for uniform distribution

[Reference link for plotting](https://www.pythonprogramming.in/change-box-color-in-boxplot.html)

$\mu_{avg} = 1/2$

```python
avg_error_array = []                     # creating a 2D array of errors between calculated and average mean in M experiments for each value of N
for i in N:
    temp = []                            # array of averages in M experiments for any given N 
    for j in range(M):
        mu = np.mean([np.random.random() for _ in range(i)])    # generating dataset, and storing it's mean in the mu array
        mu_avg = 0.5
        temp.append(abs(mu-mu_avg))
    avg_error_array.append(temp)

box = plt.boxplot(avg_error_array, patch_artist=True, showfliers=False, labels=N)                   # box plot and formatting
colors = ['blue', 'green', 'purple', 'tan', 'pink', 'red', 'yellow', 'brown', 'orange', 'cyan']
for patch, color in zip(box['boxes'], colors):
    patch.set_facecolor(color)

name = "Box plot of error for uniform distribution"
plt.title(name)
plt.xlabel("Values of N")
plt.ylabel("Error")
plt.savefig(f'../results/q5/{name}.svg')
plt.show()
```

### Generating Dataset for Gaussian distribution

$\mu_{avg} = 0$

```python
avg_error_array = []                            # creating a 2D array of errors between calculated and average mean in M experiments for each value of N
mu_avg,sigma = 0,1

for i in N:
    temp = []                                   # array of averages in M experiments for any given N 
    for j in range(M):
        mu = np.mean([np.random.normal(mu_avg, sigma) for _ in range(i)])           # generating dataset, and storing it's mean in the mu array
        temp.append(abs(mu-mu_avg))
    avg_error_array.append(temp)

box = plt.boxplot(avg_error_array, patch_artist=True, showfliers=False, labels=N)                   # box plot and formatting
colors = ['blue', 'green', 'purple', 'tan', 'pink', 'red', 'yellow', 'brown', 'orange', 'cyan']
for patch, color in zip(box['boxes'], colors):
    patch.set_facecolor(color)

name = "Box plot of error for gaussian distribution"
plt.title(name)
plt.xlabel("Values of N")
plt.ylabel("Error")
plt.savefig(f'../results/q5/{name}.svg')
plt.show()
```
