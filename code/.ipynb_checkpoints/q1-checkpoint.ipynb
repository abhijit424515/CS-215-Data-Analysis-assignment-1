```python
from matplotlib import pyplot as plt
import numpy as np

EPSILON = 1000  
LENGTH = 40
```

### Riemann Sum

```python
def riemann_sum(pdf, start, end, mu, b):
    x_arr = np.arange(start, end, 1.0/EPSILON)
    y_arr = pdf(x_arr, mu, b)
    return np.sum(y_arr)/(EPSILON)
```

### Mean function

```python
def mean(pdf, start, end, mu, b):
    xpdf = lambda x,mu,b: x*pdf(x,mu,b)
    return riemann_sum(xpdf, start, end, mu, b)
```

### RMS2 function

```python
def rms2(pdf, start, end, mu, b):
    x2pdf = lambda x,mu,b: x*x*pdf(x,mu,b)
    return riemann_sum(x2pdf, start, end, mu, b)
```

### Variance function

```python
def variance(pdf, start, end, mu, b):
    avg = mean(pdf, start, end, mu, b)
    # int_func = lambda x,mu,b: (x-avg)**2*pdf(x,mu,b)
    # return riemann_sum(int_func, start, end, mu, b)
    return rms2(pdf, start, end, mu, b) - avg**2
```

---

## Probability Distribution Function

```python
def plot_pdf(dist_f, start, end, mu, b, name):  
    x_arr = np.arange(start, end, 1/EPSILON)
    plt.plot(x_arr, dist_f(x_arr, mu, b))
    plt.xlabel("Random Variable")
    plt.ylabel("Probability Distribution Function")
    plt.title(name)
    plt.show()
```

## Cumulative Distribution Function

```python
def plot_cdf(pdf, start, end, mu, b, name):
    x_arr = np.arange(start, end, 1/EPSILON)
    #y_arr = [riemann_sum(pdf, start, i, mu, b) for i in x_arr]

    y_arr = []
    for index, elem in enumerate(x_arr):
        if(index==0):
            y_arr.append(0)
        
        else:
            # y_arr.append(y_arr[index - 1] + riemann_sum(pdf, x_arr[index - 1], elem, mu, b))
            y_arr.append(y_arr[index - 1] + pdf(elem, mu, b)/EPSILON)
    
    plt.plot(x_arr, y_arr)
    plt.xlabel("Random Variable")
    plt.ylabel("Cumulative Distribution Function")
    plt.title(name)
    plt.show()
```

## Laplace Distribution
$$P(X=x; \mu, b) = \frac{1}{2b}\cdot e^{\frac{-|x-\mu|}{b}}$$

```python
mu,b = 2,2
laplace_PDF = lambda x,m,b: (0.5/b)*np.exp((-1/b)*abs(x-m))
plot_pdf(laplace_PDF, -LENGTH, LENGTH, mu, b, "Laplace Distribution (mu=2, b=2)")
plot_cdf(laplace_PDF, -LENGTH, LENGTH, mu, b, "Laplace Distribution (mu=2, b=2)")
calc = variance(laplace_PDF, -LENGTH, LENGTH, mu, b)
exact = 2*b**2;
print(f"CALCULATED VARIANCE: {calc}")
print(f"ACTUAL VARIANCE: {exact}")
print(f"ERROR IN VARIANCE: {abs(exact-calc)}")
```

## Gumbel Distribution
$$P(X=x; \mu, \beta) = \frac{1}{\beta}\cdot e^{-\big(\frac{x-\mu}{\beta}+e^{-\frac{x-\mu}{\beta}}\big)}$$

```python
mu,b = 1,2
gumbel_PDF = lambda x,m,b: (1/b)*np.exp(-1*(((x-m)/b)+np.exp(-1*((x-m)/b))))
plot_pdf(gumbel_PDF, -LENGTH, LENGTH, mu, b, "Gumbel Distribution (mu=1, beta=2)")
plot_cdf(gumbel_PDF, -LENGTH, LENGTH, mu, b, "Gumbel Distribution (mu=1, beta=2)")
calc = variance(gumbel_PDF, -LENGTH, LENGTH, mu, b)
exact = (b**2*np.pi**2)/6;
print(f"CALCULATED VARIANCE: {calc}")
print(f"ACTUAL VARIANCE: {exact}")
print(f"ERROR IN VARIANCE: {abs(exact-calc)}")
```

## Cauchy Distribution
$$P(X=x; x_0, \gamma) = \frac{1}{\pi \gamma \big(1+(\frac{x-x_0}{\gamma})^2 \big)}$$

```python
mu,b = 0,1
cauchy_PDF = lambda x,mu,b: 1/(np.pi*b*(1+((x-mu)/b)**2))
plot_pdf(cauchy_PDF, -LENGTH, LENGTH, mu, b, "Cauchy Distribution (x0=0, gamma=1)")
plot_cdf(cauchy_PDF, -LENGTH, LENGTH, mu, b, "Cauchy Distribution (x0=0, gamma=1)")
```
